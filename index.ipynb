# widget reactivity tests with images

*This notebook demos basic forms of interactivity with images and widgets*

## setup

```python
from ipywidgets import interact
import ipywidgets as widgets
```

```python
%matplotlib inline
```

```python
from showit import image
```

## example 1

*update a square in an image*

```python
from numpy import zeros, uint8
```

```python
img = zeros((2000, 2000, 3), dtype=float)
wx = widgets.IntSlider(value=50, min=0, max=2000)
wy = widgets.IntSlider(value=50, min=0, max=2000)
@interact(x=wx, y=wy)
def f(x, y):
    img[x:x+50, y:y+50, 0] = 1
    image(img)
```

## example 2

*set a threshold on an image*

```python
from matplotlib.pyplot import imread
```

```python
oim = imread('horse.png')
w = widgets.FloatSlider(value=0.5, min=0, max=1)
@interact(t=w)
def f(t):
    img = oim.copy()
    img[img < t] = 0
    image(img)
```
